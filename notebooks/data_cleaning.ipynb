**Fixing the drug_trial_id missingness**

step 1. Extract numeric part

```python
pharma_df_clean['id_num'] = pharma_df_clean['drug_trial_id'].str.extract(r'(\d+)').astype(float)

pharma_df_clean['id_num'].head()
```

Step 2. interpolate missing values

```python
pharma_df_clean['id_num'] = pharma_df_clean['id_num'].interpolate().astype(int)

pharma_df_clean['id_num'].info()
```

Step 3. Add 'T' back to rebuild the trial ID

```python
pharma_df_clean['drug_trial_id'] = 'T' + pharma_df_clean['id_num'].astype(str)

pharma_df_clean['drug_trial_id'].isna().sum()
```

Step 4. Drop the id_num now that we no longer need it


```python
pharma_df_clean = pharma_df_clean.drop(columns = ['id_num'])

pharma_df_clean.head()
```

**Fixing missing values in the following columns**

1. drug_name
2. drug_manufacturer
3. trial_location

```python
cols = ['drug_name', 'drug_manufacturer', 'trial_location']

for item in cols:
    pharma_df_clean[item] = pharma_df_clean[item].fillna('Unknown')


pharma_df_clean[['drug_name', 'drug_manufacturer', 'trial_location']].isna().sum()
```

**Fixing missing values for drug_sample_size and dosage_mg**

While i did recommend leaving these columns as they are, since this is a very sensitive dataset that involves medicines for humans. An alternative creative imputation to facilitate research is neccesary.

My recommendation therefore, will be to find the mean of these columns based on specific groupby columns:

-  for drug_sample_size. to find the right mean to impute we will need to group by (drug_name, drug_manufacturer, and trial_phase) this will give us the closest value that we are missing. 

    - Trials of the same drug
    - From the same manufacturer
    - In the same phase

                 **Potential challenges with this plan**

If a group has only 1 or 2 rows, the “mean” is meaningless.

Example: Drug A, Manufacturer X, Phase II → only 1 trial
Missing sample size → mean = that one value
Missing dosage → mean = that one value
This is will not imputation but duplication.

- for dosage_mg. I recommend we use group by (drug_name, drug_manufacturer, trial_phase, and trial_sample_size). The resulting mean will be very specific and the closest value we can get to impute our missing value
