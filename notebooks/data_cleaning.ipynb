**Data Cleaning steps**

        Step 1. Standardizing column names 

```python
pharma_df.head()
```

```python
pharma_df_clean = pharma_df.clean_names(case_type = 'snake', strip_underscores = True)

#converted all column names into snake case and introduced underscores to link double names

pharma_df_clean.columns
```

1.b renaming columns to make them relatable and scalable

```python
pharma_df_clean.rename(
    columns = {
        'trial_id' : 'drug_trial_id',
        'company' : 'drug_manufacturer',
        'result' : 'trial_results',
        'start_date' : 'trial_start_date',
        'end_date' : 'trial_end_date',
        'participants' : 'trial_sample_size',
        'location' : 'trial_location'
    },
    inplace = True
)

#The idea is to make column names more relatable to non-technical team and also for sclability
#That is if this dataset was to be exapanded to 1000 column would the existing column make sense or would there exist
#conflicting names, For example location, is it the lab location, trial particapant location, customer location 


pharma_df_clean.columns
```

Step 2 Correcting Data types

    a. change str to numerics for trial_sample_size
    

```python
#converting trial_sample_size from string to numeric

pharma_df_clean['trial_sample_size'] = pd.to_numeric(pharma_df_clean['trial_sample_size'], errors = 'coerce')

pharma_df_clean['trial_sample_size'].dtype
```

b. change dates from str to date time format

```python
#pharma_df_clean[['trial_start_date', 'trial_end_date']]

pharma_df_clean['trial_start_date'].unique()
```

```python
pharma_df_clean['trial_end_date'].unique()

#we can see the date format here are mixed that is yyyy-mm-dd, dd-mm-yyyy, we have both '/' and '-'
#and its str note date formart
```

```python
#1. we will fix the date separation character using regex to starndardize it

pharma_df_clean[['trial_start_date', 'trial_end_date']] = pharma_df_clean[['trial_start_date', 'trial_end_date']].replace('/', '-', regex = True)

pharma_df_clean[['trial_start_date', 'trial_end_date']]
```

```python
#2. we will convert dates from str to datetime

for col in ['trial_start_date', 'trial_end_date']:
    pharma_df_clean[col] = pd.to_datetime(
        pharma_df_clean[col],
        errors = 'coerce',
        format = 'mixed',
        dayfirst = True
    )


pharma_df_clean[['trial_start_date', 'trial_end_date']] #looking at our output we can we see 

#we have cleaned our date columns
```

```python
pharma_df_clean.info()

#Our columns are in correct data types formats
```

Step 3 Fix the case type of string columns

I have noticed some are using upper, others lower and others Title. I prefer title case for all of them

```python
pharma_df_clean.head()
```

```python
pharma_df_clean['trial_phase'].unique()
```

```python
pharma_df_clean['trial_results'].unique()

#we can see several issues here, mix of case types and several spelling errors 
```

```python
#Here i will use a for loop to standardize case to title and trim leading and trailing spaces (whitespaces)

cols = ['drug_manufacturer', 'trial_results','trial_location']

for item in cols:
    pharma_df_clean[item] = pharma_df_clean[item].str.strip().str.title()


#since drug names follows scientific naming systems, leaving them in their original case is alwasy the best,
#the focus will therefore be on trimming the white spaces. 
#For the trial phase it incorporates roman numbers of which i would recommend leaving as they are in upper case
#only treaming whitespaces

for item in ['drug_name', 'trial_phase']: 
    pharma_df_clean[item] = pharma_df_clean[item].str.strip()

pharma_df_clean.head()
```

```python
pharma_df_clean['trial_results'].unique()

#after fixing the mixed cases and white spaces we can see there is still a typo! to fix this we will
#replace both 'Succce Ss', 'Sucess' with 'Success'
```

```python
pharma_df_clean['trial_results'] = pharma_df_clean['trial_results'].replace(['Sucess', 'Succce Ss'], 'Success')

pharma_df_clean['trial_results'].unique()
```
